### Homework Grades

**Question 1: Identifying Homework Columns**

```python
type(hw)
```

```python
columns_to_keep = []

# Step 2: Go through each homework number from 1 to 10
for i in range(1, 11):
    # Define the score column and max points column names
    score_col = f'Homework {i}'
    max_col = f'Homework {i} - Max Points'

    # Check if these columns are actually in the DataFrame
    if score_col in hw.columns and max_col in hw.columns:
        # Add both column names to our list if they exist
        columns_to_keep.append(score_col)
        columns_to_keep.append(max_col)

# Step 3: Make a new DataFrame using only the columns we collected
homework_df = hw[columns_to_keep]
```

```python
homework_df.head()
```

```python
#Identify and separate the columns representing homework scores and their respective maximum points.
import re

hw_col = [col for col in final_df.columns if re.match(r'Homework \d+$', col)]

hw_max = [col for col in final_df.columns if re.match(r'Homework \d+ - Max Points$', col)]
```

```python
hw_col
```

```python
hw_max
```

**Question 2: Calculating Homework Score by Total Method**

```python
total = final_df[hw_col].sum(axis=1) / final_df[hw_max].sum(axis=1)
total
```

```python
homework_df.head()
```

**Question 3: Storing Homework Score by Total Method**

```python
#Add the calculated total homework scores to the DataFrame.
final_df['HW by Total'] = total
```

**Question 4: Preparing Data for Average Calculation**

```python
#Prepare the maximum points data to align with the actual scores for element-wise division.

rename_dict = {}
for col in hw_max:
    new_col = col.split(" - Max Points")[0]
    rename_dict[col] = new_col

hw_max = final_df[hw_max].rename(columns=rename_dict)
```

```python
hw_max.head()
```

**Question 5: Calculating Homework Score by Average Method**

```python
#Calculate the average homework scores by dividing each actual score by its corresponding maximum points, summing these ratios, and then dividing by the number of assignments.
avg_hw = (final_df[hw_col] / hw_max).sum(axis=1) / len(hw_col)
avg_hw
```

**Question 6: Storing Homework Score by Average Method**

```python
#Add the calculated average homework scores to the DataFrame.
final_df['HW by Average'] = avg_hw
```

**Question 7: Determining the Final Homework Score**

```python
#Determine the final homework score for each student by taking the maximum value between the total and average methods.
final_df['Homework Score'] = final_df[['HW by Total', 'HW by Average']].max(axis=1)
```

1. Count the number of quizzes.
2. Define the maximum possible points for each quiz.
3. Compute the quiz score using the total method.
4. Compute the quiz score using the average method.
5. Determine the final quiz score by taking the maximum of the two methods.
6. Display the DataFrame to check the results

```python
#Expected Output: The DataFrame should contain an additional column 'Quiz Score' showing the final quiz score for each student, calculated as the maximum value between the total and average methods.
# 1. Filter quiz columns
quiz_len = [col for col in final_df.columns if re.match(r'Quiz \d+$', col)]

# 2. Count the number of quizzes
num_quizzes = len(quiz_len)

# 3. Define maximum points for each quiz (Assuming 10 for now)
max_points = 10

# 4. Calculate total quiz score and average quiz score
total_quiz_scores = final_df[quiz_len].sum(axis=1) / (num_quizzes * max_points)
average_quiz_scores = final_df[quiz_len].mean(axis=1) / max_points

# 5. Determine final quiz score
final_df['Quiz Score'] = pd.concat([total_quiz_scores, average_quiz_scores], axis=1).max(axis=1)

# 6. Display the DataFrame
final_df.head()
```

## Group the data to calculate final scores!

```python
weights = pd.Series(
    {
        "Exam 1 Score": 0.05,
        "Exam 2 Score": 0.10,
        "Exam 3 Score": 0.15,
        "Quiz Score": 0.30,
        "Homework Score": 0.40,
    }
)

```

1. Define the weights for each component.
2. Calculate the final score by applying the weights to each component and summing the results.
3. Calculate the ceiling score by rounding up the final score multiplied by 100.
4. Define a function to convert numeric scores to letter grades based on standard thresholds.
5. Map the ceiling scores to letter grades.
6. Store the letter grades as a categorical column in the DataFrame.
7. Display the final letter grades.

```python
#For Expected output, refer to the submissional pdf.
final_df['Final Score'] = (final_df[weights.index] * weights).sum(axis=1)

final_df['Ceiling Score'] = np.ceil(final_df['Final Score'] * 100)

def letter_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

final_df['Letter Grade'] = final_df['Ceiling Score'].apply(letter_grade)

final_df['Letter Grade'] = pd.Categorical(final_df['Letter Grade'], categories=['A', 'B', 'C', 'D', 'F'], ordered=True)

final_df[['Ceiling Score', 'Letter Grade']].head()
```
